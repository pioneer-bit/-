```python
#[batch,in_height,in_width,in_channels]  输入图片
x = tf.random.normal([1,5,5,3])
x
```

```python
# [filter_height, filter_width, in_channels, out_channels]的滤波器
```

```python
w = tf.random.normal([3,3,3,6])
```

```python
# padding [[0, 0], [pad_top,pad_bottom], [pad_left, pad_right], [0, 0]]
```

```python
out1 = tf.nn.conv2d(input=x, 
                    filters=w, 
                    strides=[1,1,1,1],
                    padding=[[0, 0], [0, 0], [0, 0],[0, 0]])
out1
```

```python
out2 = tf.nn.conv2d(input=x, 
                    filters=w, 
                    strides=1,
                    padding=[[0, 0], [0, 0], [0, 0],[0, 0]])
out2
```

```python
out3 = tf.nn.conv2d(input=x, 
                    filters=w, 
                    strides=[1,1,1,1],
                    padding='SAME') #same就是自动补零 四周补零 输出图像大小不变
out3
```

```python
cnnlayer = tf.keras.layers.Conv2D(
    filters=4, kernel_size=3)
```

```python
x
```

```python
out = cnnlayer(x)
out
```
